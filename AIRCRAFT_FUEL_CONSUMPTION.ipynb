```python
#importing modules
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
df=pd.read_csv('/content/drive/MyDrive/LSN TSAP/FUEL BURN OFF (1).csv')
```

```python
df.head()
```

```python
df.describe(include='all')
```

```python
# Set the width and height of the figure
plt.figure(figsize=(8, 6))
# Bar chart showing average fuel burn per fleet
sns.barplot(x=df['BURN OFF'], y=df['BLK HR'])
# Add label for horizontal axis
plt.xlabel("BURN OFF")
# Add label for vertical axis
plt.title("AVERAGE FUEL BURN PER BLOCK TIME")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(8, 6))
# Bar chart showing average fuel burn per aircraft
sns.barplot(x=df['BURN OFF'], y=df['AIRCRAFT'])
# Add label for horizontal axis
plt.xlabel("BURN OFF")
# Add label for vertical axis
plt.title("AVERAGE FUEL BURN PER AIRCRAFT")
```

```python
plt.title('Boxplot:Fuel Burn Off')
plt.xlabel('Burn Off')
sns.axes_style('whitegrid')
sns.boxplot(data=df, x='BURN OFF', color='blue')
```

```python
plt.title('Boxplot:Block Hours')
plt.xlabel('Block Hours')
sns.axes_style('whitegrid')
sns.boxplot(data=df, x='BLK HR', color='green');
```
